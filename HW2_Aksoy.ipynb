**Block #1: Import libraries and dataset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
anscombe=sns.load_dataset('anscombe')
print(anscombe)
```

**Block #2: Creating 2 by 2 anscombe plot**

```python
anscombe.mean()
anscombe.std()
```

```python
fig=sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=anscombe,palette="gist_rainbow_r", col_wrap=2)
#text(x, y,anscombe.mean(),anscombe.std() , fontsize=12) tried using this to add the mean and std in each box as well since the mean and std were all the same to the first decimal place
plt.show(fig)
plt.tight_layout()
#all held a seperate color through the pallete "gist_rainbow_r"
#was able to wrap the graphs for a 2*2 
#labelled x and y axis 
```

**Saving figuring with 125 dpi to keep the anscombe model clear as a .png file**

```python
fig.savefig('fig_anscombe',dpi=125)
```
